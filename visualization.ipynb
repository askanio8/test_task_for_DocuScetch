```python
import pandas as pd
from drawing.drawing import PlotDrawer
from IPython.display import Image, display

# Чтение JSON-файла
file_url = "https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json"
df = pd.read_json(file_url)
```

```python
# Проверка данных
# Проверка, что модель находила все углы
df['temp_column'] = df['gt_corners'] - df['rb_corners']
non_zero_count = df['temp_column'].astype(bool).sum()
print("Количество ненулевых элементов в столбце:", non_zero_count)
df = df.drop(columns=['temp_column'])

# Проверка, что все значения в столбце 'max' больше значений в столбце 'mean'
max_greater_than_mean = (df['max'] > df['mean']).all()
print(f"Все значения в столбце 'max' больше значений в столбце 'mean': {max_greater_than_mean}")
# Проверка, что все значения в столбце 'min' меньше значений в столбце 'mean'
min_less_than_mean = (df['min'] < df['mean']).all()
print(f"Все значения в столбце 'min' меньше значений в столбце 'mean': {min_less_than_mean}")

# Проверка, что все значения в столбце 'floor_max' больше значений в столбце 'floor_mean'
max_greater_than_mean = (df['floor_max'] > df['floor_mean']).all()
print(f"Все значения в столбце 'floor_max' больше значений в столбце 'floor_mean': {max_greater_than_mean}")
# Проверка, что все значения в столбце 'floor_min' меньше значений в столбце 'floor_mean'
min_less_than_mean = (df['floor_min'] < df['floor_mean']).all()
print(f"Все значения в столбце 'floor_min' меньше значений в столбце 'floor_mean': {min_less_than_mean}")

# Проверка, что все значения в столбце 'ceiling_max' больше значений в столбце 'ceiling_mean'
max_greater_than_mean = (df['ceiling_max'] > df['ceiling_mean']).all()
print(f"Все значения в столбце 'ceiling_max' больше значений в столбце 'ceiling_mean': {max_greater_than_mean}")
# Проверка, что все значения в столбце 'ceiling_min' меньше значений в столбце 'ceiling_mean'
min_less_than_mean = (df['ceiling_min'] < df['ceiling_mean']).all()
print(f"Все значения в столбце 'ceiling_min' меньше значений в столбце 'ceiling_mean': {min_less_than_mean}")
```
